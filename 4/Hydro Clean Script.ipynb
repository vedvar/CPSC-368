```python
from datetime import datetime
import pandas as pd
from pandas import Timestamp
```

```python
data = pd.read_csv('Middle_Vernon_Creek_Hydro_Data.csv')
data
```

```python
for i in range(len(data)):
    
    data.at[i,'LocalDay'] = Timestamp(data.at[i,' Date_Time_UTC']).to_pydatetime().day
    data.at[i,'LocalMonth'] = Timestamp(data.at[i,' Date_Time_UTC']).to_pydatetime().month
    data.at[i,'LocalYear'] = Timestamp(data.at[i,' Date_Time_UTC']).to_pydatetime().year
    
data
```

```python
data['datetime'] = pd.to_datetime(data[' Date_Time_UTC'])

# Extract year and month from datetime and create new columns
data['Year'] = data['datetime'].dt.year
data['Month'] = data['datetime'].dt.month
data['Day'] = data['datetime'].dt.day

# Calculate the average value for each unique (year, month) combination
monthly_avg = data.groupby(['Year', 'Month','Day'])[' Value'].mean()
```

```python
temp = pd.DataFrame(monthly_avg).reset_index(drop=False)
temp
```

```python
temp['Location'] = data.at[0,data.columns[0]]
temp['Location_Name'] = data.at[0,data.columns[1]]
temp['Latitude'] = data.at[0,data.columns[3]]
temp['Longitude'] = data.at[0,data.columns[4]]
temp['Unit'] = data.at[0,data.columns[7]]
```
